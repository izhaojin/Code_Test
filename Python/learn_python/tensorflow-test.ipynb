```python
import numba as nb
import numpy as np
import matplotlib.pyplot as plt
```

```python
@nb.vectorize
def triangle_wave(A:float,T:float,x:float)->float:
    # if x<0:
    #     return -triangle_wave(-x)
    x = x%T
    if x<T/2:
        return x*(2*A/T)
    elif x<T/2*3:
        return -(x-T)*(2*A/T)
    else:
        return x*(2*A/T)
```

```python
xv = np.arange(-10,10,0.1)
A = 2
T = 4
yv = triangle_wave(A,T,xv)
```

```python
plt.plot(xv,yv)
```
